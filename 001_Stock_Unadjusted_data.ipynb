# Price Before Split and Price After Split- Remove the effect of Stock Split in Price

```python
print('Price After Split: ',df.loc["2006-09-08"].Close)
print('Price Before Split:',((df.loc["2006-09-08"].Close)*(df.loc["2006-09-08"]['Stock Splits'])))
```

```python
df['Stock Splits'].replace(to_replace=0,value=1,inplace=True)
df.head()
```

```python
df["Multiplier"]=df.sort_index(ascending=False)['Stock Splits'].cumprod()
df.head()
```

```python
df['Unadj Closed']=df["Close"]*df["Multiplier"]
df.head()
```

```python
df[["Close", 'Unadj Closed']].plot(figsize=(12,8));
```

# Unadjusted dividends

```python
df["Unadj Div"]=df.Dividends*df.Multiplier
df[df.Dividends!=0]
```

```python
df[df.Dividends!=0][['Unadj Div','Stock Splits']].plot();
```

# UnAdjusted Trading Volume

```python
df['unadj Vol']=df.Volume/df.Multiplier
df['unadj Vol'].plot()
```

```python
df[df.Dividends!=0][['unadj Vol','Unadj Closed']].plot()
```
