```python
count=1
for symbol in symbols:
  try:
    info=pd.Series(yf.Ticker(ticker=symbol).get_info()).to_frame().T
    cs=pd.concat([cs,info])
    print(count,end='\r')
    # print('{}'.format(symbol))
    count+=1
  except Exception as e:
    print('{} not found'.format(symbol))
print('Donwload complete')
```
