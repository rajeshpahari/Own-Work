```python
import pandas as pd
```

```python
url='https://www.interactivebrokers.com/en/index.php?f=2222&exch=nse&showcategories=STK&p=&cc=&limit=100&page={}'
```

```python
pd.read_html(url.format(1))[2]
```

```python
pages=22
```

```python
df=pd.DataFrame()
df
```

```python
for i in range(1,pages+1):
  table=pd.read_html(url.format(i))[2]
  df=pd.concat([df,table])
```

```python
df.head()
```

```python
df.index
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.index
```

```python
df.head()
```

```python
import yfinance as yf
```

```python
symbols=df.Symbol+'.NS'
symbols= symbols.to_list()
# symbol
```

```python
count=0
cs=pd.DataFrame()
for symbol in symbols:
  try:
    
    data=pd.Series(yf.Ticker(ticker=symbol).get_info()).to_frame().T
    cs=pd.concat([cs,data])
    count+=1
    print(count,'...',symbol)
  except Exception as e:
    print('{} is not available'.format(symbol))

print('download is completed')
```
