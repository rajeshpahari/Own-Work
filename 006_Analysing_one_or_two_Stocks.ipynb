```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
pd.options.display.float_format='{:.4f}'.format
plt.style.use('seaborn-v0_8-dark')
```

```python
# print(plt.style.available)
```

```python
start ='2022-1-1'
end='2023-4-3'
symbol=['LTIM.NS','EXIDEIND.NS']
```

```python
df=yf.download(symbol,start,end)
df
```

```python
close=df.Close.copy()
close
```

```python
close['LTIM.NS'].dropna().plot(figsize=(10,6),fontsize=13)
plt.legend(fontsize=13)
plt.show()
```

```python
close.div(close.iloc[0]).mul(100)
```

```python
#This chart is in normalized data
close.div(close.iloc[0]).mul(100).dropna().plot(figsize=(10,6),fontsize=13,logy=False)
plt.legend(fontsize=13)
plt.show()
```
